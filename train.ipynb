```python
from ultralytics import YOLO
import operator
import numpy as np
# Load a model
model = YOLO('/content/runs/classify/train3/weights/best.pt')  # load a custom model

def get_labels(path,n):
    # n = number of labels to be predicted 
    results = model(path)
    class_labels = results[0].names
    probs = results[0].probs.data.tolist()
    total_class = len(class_labels)
    if n>total_class:
        n = total_class
        
    # result = dict(zip(class_labels.values(), probs))
    # sorted_result = dict(sorted(result.items(), key=operator.itemgetter(1), reverse= True)[:n])
    sorted_result = class_labels[np.argmax(probs)]
    return sorted_result

path = 'https://m.media-amazon.com/images/I/A1lwl9BzXGL._AC_UY1100_.jpg'
print(get_labels(path,n=10))
```

```python
def create_visualization(base_path):
    import os
    import ipyplot
    from PIL import Image
    result = {}
    for image_name in os.listdir(base_path):
        image_paths = os.path.join(base_path, image_name)
        label = get_labels(image_paths, 10)
        result[image_paths] = label
        
    labels = list(result.values())
    images  = list(result.keys())
    images_pil = [Image.open(image) for image in images]
    return labels ,images_pil

base_path = '/content/drive/MyDrive/Notebooks/visualize'
labels, images_pil = create_visualization(base_path)  
```

```python
import ipyplot
ipyplot.plot_class_tabs(images_pil, labels, img_width = 150)    
```
